```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0,9.0)
import seaborn as sns
```

```python
data = pd.read_csv("Salary_Data.csv")
```

```python
X = data.iloc[:,0]
Y = data.iloc[:,1]
```

```python
plt.scatter(X,Y)
```

```python
m = 0 #slope of X
c = 0 #intersept 

l = 0.0001 #learning rate at 0.0001
n = float(len(X)) #lengh of X
epochs = 1000
for i in range(epochs):
    Y_pred = m*X + c #current predicted  value of Y
    D_m = (-2/n)*sum(X*(Y-Y_pred)) #derivative of m
    D_c = (-2/n)*sum(Y-Y_pred) #derivative of c
    
    m = m - l*D_m #update m
    c = c - l*D_c #update c
print(m,c)
```

```python
Y_pred = m*X+c

plt.scatter(X,Y)
plt.plot(X, Y_pred)
```

```python
plt.scatter(X,Y)
plt.plot([min(X),max(X)], [min(Y_pred),max(Y_pred)], color = 'red')
```
